**Activida**: Primero vamos a crear una función que genera una lista de números con distribución gaussiana. Para eso haremos uso de la función `gauss` de la librería `random`. La función debe tomar tres parámetros: la cantidad de números a generar y el promedio y desviación estándar de la distribución. Debe devolver una lista con esos números.

```python

def distribucion(n, mu, sigma):
    lista = []
    for i in range(n):
        lista.append(rd.gauss(mu,sigma))
    
    
    return lista

```

Unir los códigos para crear una función que tome un lista y retorna una lista de tuplas con las frecuencias de cada elemento

```python
def frecuencias(lista):
    frec = {}
    for item in lista:
        if item in frec:
            frec[item] += 1
        else:
            frec[item] = 1
            
    lista_frecuencias = []
    for key in frec:
        lista_frecuencias.append((key, frec[key]))
    
    return lista_frecuencias
```

**Actividad**: ahora nos interesa poder ordenar la lista según la frecuencia de cada tupla. Hay funciones de python que nos permiten hacer esto, pero vamos a hacer una actividad para crear un código que ordene la lista sin usar ninguna función de Python más que el método `append`. Hay muchas formas de hacerlo, acá vamos a proponer una forma. Vamos a recorrer la lista sucesivas veces buscando el elemento cuya frecuencia sea la menor. Una vez encontrado el elemento con menor frecuencia lo agregamos como elemento de la lista ordenada y lo borramos de la lista original. Repetimos este proceso hasta que nuestra lista original quede vacía.

```python
ordenada = []
while len(lista_frec)>0:
    index = 0
    min_index = 0
    for item in lista_frec :
        if item[1] < lista_frec[min_index][1]:
            min_index = index
            index += 1
        ordenada.append(lista_frec[min_index])
        del lista_frec[min_index]
        
ordenada
```

Converitir el código anterior en una función para usarla más adelante, pero comenzar copiando la lista y trabajar sobre la copia para no perderla luego de procesarla

```python
def ordenar_frecuencias(lista_frec):
    copia = lista_frec.copy()
   
    ordenada = []
    while len(copia)>0:
        index = 0
        min_index = 0
        for item in copia:
            if item[1] < copia[min_index][1]:
                min_index = index
            index += 1
        ordenada.append(copia[min_index])
        del copia[min_index]

    

    return ordenada
```

**Actividad**: borrar de la lista `palabras` los tokens que no sean palabras en el formato necesario

```python
indices_a_borrar= []
for index , palabra in enumerate(palabras):
    if not palabra.isalpha():
        indices_a_borrar.append(index)

        
for indice in indices_a_borrar[::-1]:
    del palabras[indice]
```
